```python
# =========================================================
# PERTEMUAN 3 - DATA SCIENCE
# Nama : Amin Siddik Rangkuti
# NIM  : 220401010124
# =========================================================

# =========================================================
# IMPORT LIBRARY
# =========================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# =========================================================
# SEL 1 : MENAMPILKAN IDENTITAS
# =========================================================
print("======================================")
print("PENGANTAR DATA SCIENCE - PERTEMUAN 3")
print("======================================")
print("Nama : Amin Siddik Rangkuti")
print("NIM  : 220401010124")
print("======================================")
```

```python
# =========================================================
# SEL 2 : MEMBUAT DATAFRAME
# =========================================================
data = {
    "Nama": ["Andi", "Budi", "Citra", "Dewi", "Eko"],
    "Nilai": [85, 90, 78, 88, 95],
    "Kehadiran": [90, 95, 80, 85, 100]
}

df = pd.DataFrame(data)

print("\nDATAFRAME MAHASISWA")
print(df)
```

```python
# =========================================================
# SEL 3 : INFORMASI DATAFRAME
# =========================================================
print("\nINFORMASI DATAFRAME")
print(df.info())
```

```python
# =========================================================
# SEL 4 : STATISTIK DESKRIPTIF
# =========================================================
print("\nSTATISTIK DATA")
print(df.describe())
```

```python
# =========================================================
# SEL 5 : FILTER DATA
# =========================================================
print("\nMAHASISWA DENGAN NILAI > 85")
print(df[df["Nilai"] > 85])
```

```python
# =========================================================
# SEL 6 : MENAMBAHKAN KOLOM STATUS
# =========================================================
df["Status"] = ["Lulus" if n >= 75 else "Tidak Lulus"
                for n in df["Nilai"]]

print("\nDATAFRAME SETELAH TAMBAH KOLOM STATUS")
print(df)
```

```python
# =========================================================
# SEL 7 : MENGHITUNG RATA-RATA
# =========================================================
rata_nilai = df["Nilai"].mean()
rata_kehadiran = df["Kehadiran"].mean()

print("\nRATA-RATA NILAI :", rata_nilai)
print("RATA-RATA KEHADIRAN :", rata_kehadiran)
```

```python
# =========================================================
# SEL 8 : NILAI TERTINGGI DAN TERENDAH
# =========================================================
print("\nNILAI TERTINGGI :", df["Nilai"].max())
print("NILAI TERENDAH  :", df["Nilai"].min())
```

```python
# =========================================================
# SEL 9 : SORTING DATA
# =========================================================
print("\nDATA DIURUTKAN BERDASARKAN NILAI")
print(df.sort_values(by="Nilai", ascending=False))
```

```python
# =========================================================
# SEL 10 : NUMPY ARRAY
# =========================================================
angka = np.array([10, 20, 30, 40, 50])

print("\nNUMPY ARRAY")
print(angka)

print("\nARRAY DIKALI 2")
print(angka * 2)
```

```python
# =========================================================
# SEL 11 : VISUALISASI DATA
# =========================================================
plt.figure(figsize=(8,5))

plt.bar(df["Nama"], df["Nilai"])

plt.title("Grafik Nilai Mahasiswa")
plt.xlabel("Nama")
plt.ylabel("Nilai")

plt.show()
```
