```python
# =========================================================
# PERTEMUAN 3 - PENGANTAR DATA SCIENCE
# Nama : Amin Siddik Rangkuti
# NIM  : 220401010124
# =========================================================

# =========================================================
# IMPORT LIBRARY
# =========================================================
import pandas as pd
import numpy as np
```

```python
# =========================================================
# SEL 1 : MEMBUAT DATAFRAME
# =========================================================
data = {
    "Nama": ["Andi", "Budi", "Citra", "Dewi", "Eko"],
    "Umur": [20, 21, 19, 22, 20],
    "Nilai": [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)

print("DATAFRAME MAHASISWA")
print(df)
```

```python
# =========================================================
# SEL 2 : MENAMPILKAN INFORMASI DATA
# =========================================================
print("\nINFORMASI DATA")
print(df.info())
```

```python
# =========================================================
# SEL 3 : MENAMPILKAN STATISTIK DATA
# =========================================================
print("\nSTATISTIK DATA")
print(df.describe())
```

```python
# =========================================================
# SEL 4 : MENAMPILKAN KOLOM TERTENTU
# =========================================================
print("\nKOLOM NAMA")
print(df["Nama"])

print("\nKOLOM NILAI")
print(df["Nilai"])
```

```python
# =========================================================
# SEL 5 : FILTER DATA
# =========================================================
print("\nMAHASISWA DENGAN NILAI > 85")
print(df[df["Nilai"] > 85])
```

```python
# =========================================================
# SEL 6 : MENAMBAHKAN KOLOM BARU
# =========================================================
df["Status"] = ["Lulus", "Lulus", "Lulus", "Lulus", "Lulus"]

print("\nDATAFRAME SETELAH TAMBAH KOLOM")
print(df)
```

```python
# =========================================================
# SEL 7 : RATA-RATA NILAI
# =========================================================
rata_rata = df["Nilai"].mean()

print("\nRATA-RATA NILAI")
print(rata_rata)
```

```python
# =========================================================
# SEL 8 : NILAI TERTINGGI DAN TERENDAH
# =========================================================
print("\nNILAI TERTINGGI")
print(df["Nilai"].max())

print("\nNILAI TERENDAH")
print(df["Nilai"].min())
```

```python
# =========================================================
# SEL 9 : SORTING DATA
# =========================================================
print("\nDATA DIURUTKAN BERDASARKAN NILAI")
print(df.sort_values(by="Nilai", ascending=False))
```

```python
# =========================================================
# SEL 10 : NUMPY ARRAY
# =========================================================
array_angka = np.array([10, 20, 30, 40, 50])

print("\nNUMPY ARRAY")
print(array_angka)

print("\nHASIL PERKALIAN ARRAY x2")
print(array_angka * 2)
```
